```python
import pandas as pd
import numpy as np
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python

excel = pd.ExcelFile('train_data.xlsx') #aceesing the excel
data = pd.read_excel(excel, 'Sheet1')
```

```python
excel = pd.ExcelFile('test_data.xlsx') #aceesing the excel
data2 = pd.read_excel(excel, 'Sheet1')
```

```python
t5_merged_data = pd.concat([data, data2], axis=0)

# Remove class labels
t5_features = t5_merged_data.drop('embed_9', axis=1)

k_t5 = 3

kmeans_t5 = KMeans(n_clusters=k_t5, random_state=42)
clusters_t5 = kmeans_t5.fit_predict(t5_features)

from sklearn.preprocessing import StandardScaler

# Standardize or normalize features
scaler = StandardScaler()
t5_features_scaled = scaler.fit_transform(t5_features)

# Add the cluster labels back to the DataFrame
t5_merged_data['cluster'] = clusters_t5

# Now 't5_merged_data' contains the original features, and a new column 'cluster' with the assigned cluster labels

# Display the clustered dataset
print(t5_merged_data)
```
